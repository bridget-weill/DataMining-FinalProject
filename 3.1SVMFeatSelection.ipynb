### Import and Read in Data

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
data = pd.read_csv('pipeline3_1.csv', header=0)
data.head()
```

### Split Features and Target

```python
X = data.iloc[:, :74]
y = data.iloc[:, 74:75]
```

### Select Features Using F Classif 
Need to use F CLassif since we have negative values in our data.  This is due to the order of pipeline 1.

```python
# selecting the best features
# ref: https://medium.com/analytics-vidhya/preprocessing-data-for-predicting-online-shoppers-purchasing-intention-ml-ba78186b7e85

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

bestfeatures = SelectKBest(f_classif, k=2)

fit = bestfeatures.fit(X, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Feature', 'Score']
print(featureScores.nlargest(30, 'Score'))
```
